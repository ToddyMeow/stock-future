# HAB Bars Full History By Symbol

输入一个品种代码，例如 `CU`、`IF`、`AU`，读取 `hab_bars.csv`，并绘制这个品种的全历史 `open/high/low/close` 曲线。

```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd

repo_root = Path.cwd()
if repo_root.name == "note":
    repo_root = repo_root.parent

hab_bars_path = repo_root / "data" / "cache" / "normalized" / "hab_bars.csv"
hab_bars_path
```

```python
bars = pd.read_csv(hab_bars_path, parse_dates=["date"])
bars = bars.sort_values(["date", "symbol"]).reset_index(drop=True)

print(f"date range: {bars['date'].min().date()} -> {bars['date'].max().date()}")
print(f"available symbols: {bars['symbol'].nunique():,}")
sorted(bars['symbol'].unique())[:20]
```

```python
# 修改这里的品种代码
symbol = "CU"
```

```python
symbol = symbol.strip().upper()
symbol_df = bars.loc[bars["symbol"] == symbol].copy()

if symbol_df.empty:
    available = ", ".join(sorted(bars["symbol"].unique())[:50])
    raise ValueError(f"symbol {symbol!r} not found in hab bars. examples: {available}")

print(f"symbol: {symbol}")
print(f"rows: {len(symbol_df):,}")
print(f"date range: {symbol_df['date'].min().date()} -> {symbol_df['date'].max().date()}")
symbol_df.head()
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(symbol_df["date"], symbol_df["open"], label=f"{symbol} open", linewidth=1.4)
ax.plot(symbol_df["date"], symbol_df["high"], label=f"{symbol} high", linewidth=1.4)
ax.plot(symbol_df["date"], symbol_df["low"], label=f"{symbol} low", linewidth=1.4)
ax.plot(symbol_df["date"], symbol_df["close"], label=f"{symbol} close", linewidth=2.0)

ax.set_title(f"{symbol} OHLC - Full History")
ax.set_xlabel("Date")
ax.set_ylabel("Price")
ax.grid(True, alpha=0.25)
ax.legend()
fig.autofmt_xdate()
plt.show()
```
